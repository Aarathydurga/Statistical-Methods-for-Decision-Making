# Outlier Treatment

```python
# Define a function to identify outliers using IQR
def detect_outliers(col):
    q1 = col.quantile(0.25)
    q3 = col.quantile(0.75)
    iqr = q3 - q1
    return col[(col < q1 - 1.5 * iqr) | (col > q3 + 1.5 * iqr)]

# Find the columns with outliers
outlier_cols = []
for col in df.columns:
    if np.issubdtype(df[col].dtype, np.number):
        outliers = detect_outliers(df[col])
        if len(outliers) > 0:
            outlier_cols.append(col)

# Create a boxplot for the columns with outliers
fig, ax = plt.subplots(figsize=(10, 5))
df[outlier_cols].boxplot(ax=ax, vert=False)
ax.set_title('Boxplot of Columns with Outliers')
plt.show()
```

To remove outliers


```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range    
```

```python
lr,ur=remove_outlier(df['cc_limit'])
df['cc_limit']=np.where(df['cc_limit']>ur,ur,df['cc_limit'])
df['cc_limit']=np.where(df['cc_limit']<lr,lr,df['cc_limit'])

lr,ur=remove_outlier(df['avg_spends_l3m'])
df['avg_spends_l3m']=np.where(df['avg_spends_l3m']>ur,ur,df['avg_spends_l3m'])
df['avg_spends_l3m']=np.where(df['avg_spends_l3m']<lr,lr,df['avg_spends_l3m'])


lr,ur=remove_outlier(df['annual_income_at_source'])
df['annual_income_at_source']=np.where(df['annual_income_at_source']>ur,ur,df['annual_income_at_source'])
df['annual_income_at_source']=np.where(df['annual_income_at_source']<lr,lr,df['annual_income_at_source'])







```
